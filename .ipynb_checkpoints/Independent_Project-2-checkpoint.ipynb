The session data file constitute 1535 rows and 5 columns. Each row represents the start and end time of the player identified by hashedEmail. 
Issued in this data set is similar to the player data:

- The hashedEmail is very long but it is the unique identification of each player.
- While there are 1535 rows, a lot of them are records of same players. Therefore, these rows need to be combined according to their hashedEmail.

```r
sessions_summary <- session |> 
  group_by(hashedEmail) |> 
  summarise(across(everything(), ~ if(is.numeric(.)) mean(., na.rm = TRUE) else first(.)))

# Merge with players dataset
merged_data <- inner_join(players, sessions_summary, by = "hashedEmail")

# View the merged dataset
nrow(merged_data)
merged_data
```

```r

# Calculate average age
Avg_age <- players |> 
  summarise(Avg_age = mean(Age, na.rm = TRUE)) 

# Calculate average played hours (Fixed typo in 'payers')
Avg_hours <- players |> 
  summarise(Avg_hours = mean(played_hours, na.rm = TRUE))

# Filter gender to only include Male and Female (Fixed mutate issue)
Modi_gender <- players |> 
  filter(gender %in% c("Male", "Female"))

# Print results
Avg_age
Avg_hours

avg_hours_by_experience <- merged_data |> 
  group_by(experience) |> 
  summarise(Avg_played_hours = mean(played_hours, na.rm = TRUE))
avg_hours_by_experience


avg_ages_by_experience <- merged_data |> 
  group_by(experience) |> 
  summarise(Avg_age = mean(Age, na.rm = TRUE))
avg_ages_by_experience
```

Lets visualize the relationship of played hour and age of individuals and find out the clusters of experience group.

```r
graph1 <- ggplot(merged_data, aes(x = Age, y = played_hours, color = experience)) +
  geom_point(alpha = 0.6) +  # Scatter plot with transparency
  labs(title = "Played Hours vs. Age by Experience Level",
       x = "Age",
       y = "Played Hours",
       color = "Experience Level") 
graph1
```

From obsrving the graph, a lot of data points are accmulated at the bottom. This is likely due to the wide range of hours recorded in played_hours. In order to solve this problem, we can convert the played_hour using log function.

```r

 # Should be > 0

# Plot with log transformation and jittering
graph2 <- ggplot(merged_data, aes(x = Age, y = played_hours, color = experience)) +
  geom_point(alpha = 0.6, position = position_jitter(width = 0.5, height = 0)) +  
  scale_y_log10() +  # Log-transform Y-axis to spread out small values
  scale_color_viridis_d() +  # Better color scheme for accessibility
  labs(title = "Played Hours vs. Age by Experience Level (Log Scale)",
       x = "Age",
       y = "Log(Played Hours)",
       color = "Experience Level") + 
  theme_minimal()
graph2
```

The graph seems better than the previous one. The points are more distributed across the graph. Also by filtering out the extreme time values such as those larger than 10000, it have made the graph more clean.
